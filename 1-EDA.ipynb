```python
import os, random, cv2
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_folder = 'brain_tumor_dataset/'
dataset_dividido_folder = 'brain_tumor_dataset_dividido/'

treino_folder = 'treino/'
teste_folder = 'teste/'

yes = 'yes/'
no = 'no/'

path = os.listdir(dataset_folder + yes)
sample = random.sample(path,6)
f,ax = plt.subplots(2,3) 
for i in range(0,6):
    img = cv2.imread(dataset_folder + yes + sample[i])
    ax[i//3,i%3].imshow(img)
    ax[i//3,i%3].axis('off')
f.suptitle('Rais X com tumores')
plt.show()
```

```python
path = os.listdir(dataset_folder + no)
sample = random.sample(path,6)
f,ax = plt.subplots(2,3) 
for i in range(0,6):
    img = cv2.imread(dataset_folder + no + sample[i])
    ax[i//3,i%3].imshow(img)
    ax[i//3,i%3].axis('off')
f.suptitle('Rais X com tumores')
plt.show()
```

```python
dir_yes = os.listdir(dataset_folder + yes)
dir_no = os.listdir(dataset_folder + no)
fig_dims = (5, 5)
fig, ax = plt.subplots(figsize=fig_dims)
sns.set_style('whitegrid')
grafico = sns.barplot(ax = ax, x=['Yes','No'], y=[len(dir_yes),len(dir_no)])
grafico.set_title('Dados Dataset Inicial')
plt.show()
acc_aleatoria = len(dir_yes)/(len(dir_yes)+len(dir_no))
print('Inicialmente, começamos com uma percentagem de', acc_aleatoria, '% imagens com positivo para tumor cerebral!')
print('Ou seja, uma boa baseline de comparação / avaliação da nossa accuracy, será:',acc_aleatoria,'%')
```

```python
dataset_dividido_folder_treino = dataset_dividido_folder + treino_folder
dataset_dividido_folder_teste = dataset_dividido_folder + teste_folder

dir_train_yes = os.listdir(dataset_dividido_folder_treino + yes)
dir_train_no = os.listdir(dataset_dividido_folder_treino + no)
dir_test_yes = os.listdir(dataset_dividido_folder_teste + yes)
dir_test_no = os.listdir(dataset_dividido_folder_teste + no)
fig_dims = (5, 5)
fig, ax = plt.subplots(figsize=fig_dims)
sns.set_style('whitegrid')
grafico = sns.barplot(ax = ax, x=['Train Yes','Train No','Test Yes','Test No'], y=[len(dir_train_yes),len(dir_train_no),len(dir_test_yes),len(dir_test_no)])
grafico.set_title('Dados Dataset Dividido')
plt.show()
```
